## End to End Deep Learning Projrct

* Create a `ResNet50` Model for classifying Imagenet image
* Create a Flask App to upload the image (Html)
* Predict From the Flask App with the uploaded images

```python
from tensorflow.keras.applications.resnet50 import ResNet50
```

```python
model = ResNet50(weights='imagenet') #downloads all the default weights
```

```python
model.summary()
```

```python
model.save('resnet50.h5')
```

```python
from tensorflow.keras.applications.vgg19 import VGG19

model1 = VGG19(weights="imagenet")
```

```python
model1.summary()
```

```python
model1.save('vgg19.h5')
```

The `VGG19` model has many more parameters than the `ResNet50` model, but the later is more optimized to give better accuracy.
